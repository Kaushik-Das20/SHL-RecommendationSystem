```python
import pandas as pd
import requests
import time
```

```python
test_df = pd.read_excel("Gen_AI Dataset.xlsx", sheet_name=1)

print(f"Total queries: {len(test_df)}")
print(f"Columns: {list(test_df.columns)}")
```

```python
API_URL = "http://localhost:8000"

def get_predictions(query, top_k=10):
    try:
        response = requests.post(
            f"{API_URL}/recommend",
            json={"query": query, "top_k": top_k},
            timeout=60
        )
        if response.status_code == 200:
            data = response.json()
            urls = [a["url"] for a in data["recommended_assessments"]]
            return urls
        else:
            print(f"❌ API Error: {response.status_code}")
            return []
    except Exception as e:
        print(f" Error: {e}")
        return []
```

```python
all_predictions = []

for i, row in test_df.iterrows():
    query = row["Query"]
    
    #print(f"Processing Query {i+1}/{len(test_df)}: {query[:60]}...")
    
    predicted_urls = get_predictions(query, top_k=10)
    print(f" {len(predicted_urls)} recommendations")
    
    for url in predicted_urls:
        all_predictions.append({
            "Query": query,
            "Assessment_url": url
        })
    
    time.sleep(1)
```

```python
predictions_df = pd.DataFrame(all_predictions)

predictions_df.to_csv("kaushik_das.csv", index=False)

print(f"\nTotal rows: {len(predictions_df)}")
print(f"Unique queries: {predictions_df['Query'].nunique()}")
print(f"\nPreview:")
predictions_df.head(10)
```
